OUTLIER DETECTION AND REMOVAL

```python
#Plot the distribution plots for the 'Fare' feature
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(10,5))
plt.subplot(1,1,1)
sns.distplot(data1['Fare'])
plt.show()
```

```python
#boxplot for 'Fare' Feature
plt.boxplot(data1['Fare'])
plt.show()
```

```python
#Finding IQR
percentile25 = data1['Fare'].quantile(0.25)
percentile75 = data1['Fare'].quantile(0.75)
iqr=percentile75-percentile25

#finding upper and lower limits
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr

#finding outliers
data1[data1['Fare'] > upper_limit]
data1[data1['Fare'] < lower_limit]

#trimming outliers
new_data1 = data1[data1['Fare'] < upper_limit]
```

```python
# Comparing plots after trimmimg
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(data1['Fare'])
plt.subplot(2,2,2)
sns.boxplot(data1['Fare'])
plt.subplot(2,2,3)
sns.distplot(new_data1['Fare'])
plt.subplot(2,2,4)
sns.boxplot(new_data1['Fare'])
plt.show()
```
