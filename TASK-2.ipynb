MATPLOTLIB

```python
#Visualization with matplotlib

import matplotlib.pyplot as plt
x=[10,23,34,35,44]
y=[10,20,30,40,50]
plt.plot(x,y)
plt.show()
```

```python
#Plot with title,labels

plt.title('Customized Plot')
plt.plot(x, y, color='green', linestyle='dotted', marker='^')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#Plot with multiple datasets

import matplotlib.pyplot as plt

x=[10,23,34,35,44]
y1=[10,20,30,40,50]
y2 =[12,24,35,42,50]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.title('Multiple Lines on One Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc="upper right")
plt.show()
```

Types of plots

Scatter plot

```python
import matplotlib.pyplot as plt
x=[10,23,34,35,44]
y=[10,20,30,40,50]
plt.scatter(x, y, color='green', marker='^', label='Plotted Points')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

Bar plot

```python
Subjects=['Physics','Chemistry','Mathematics']
scores=[78,65,92]
plt.bar(Subjects,scores,color="green")
plt.xlabel('SUBJECTS')
plt.ylabel('MARKS')
plt.show()
```

Histogram

```python
roll= [1,1,2,3,3,3,4,4,5,5,5,5]
plt.hist(roll,bins=5,edgecolor='k',color='skyblue')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

Piechart

```python
Subjects=['Physics','Chemistry','Mathematics']
scores=[78,65,92]
plt.pie(scores,labels=Subjects,autopct='%1.1f%%')
plt.show()
```

Subplots and Layouts

```python
#Subplots: Dividing a figure into multiple subplots with plt.subplot()
import matplotlib.pyplot as plt

# Create a figure with 2x2 subplots
plt.figure(figsize=(10, 6))  # Set the overall figure size
plt.subplot(2, 2, 1)  # The first subplot
plt.plot([1, 2, 3, 4], [10, 15, 13, 18])
plt.title('Subplot 1')

plt.subplot(2, 2, 2)  # The second subplot
plt.scatter([1, 2, 3, 4], [10, 15, 13, 18], color='red')
plt.title('Subplot 2')

plt.subplot(2, 2, 3)  # The third subplot
plt.bar(['A', 'B', 'C'], [15, 30, 25], color='green')
plt.title('Subplot 3')
plt.subplot(2, 2, 4)  # The fourth subplot
plt.hist([1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5], bins=5, edgecolor='k', color='skyblue')
plt.title('Subplot 4')

plt.suptitle('Subplots Example')  # Set a title for the entire figure
plt.tight_layout()  # Ensure proper spacing
plt.show()
```

Advanced Plot Types

```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

# Customize the plot
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')

plt.show()
```

Seaborn

```python
#Data Visualization with Seaborn

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Barplot

```python
data = sns.load_dataset("tips")
sns.barplot(data=data, x="day", y="total_bill", ci=None, palette="Set2")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.title("Average Total Bill by Day")
plt.show()
```

Line Plot

```python
data = sns.load_dataset("flights")
sns.lineplot(data=data, x="year", y="passengers")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.title("Line Plot of Number of Passengers Over Time")
plt.show()
```

Scatter Plot

```python
data = sns.load_dataset("iris")
sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="species", palette="viridis")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatter Plot of Sepal Length vs. Sepal Width")
plt.show()
```

Histogram

```python
sns.histplot(data=data, x="sepal_length", kde=True, color="skyblue")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")
plt.show()
```

Box plot

```python
sns.boxplot(data=data, x="species", y="petal_length", palette="Set3")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.title("Box Plot of Petal Length by Species")
plt.show()
```
